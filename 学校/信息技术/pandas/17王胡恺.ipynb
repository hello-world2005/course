```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('超市销售数据.csv', encoding = 'GBK')
```

#### 最受欢迎的商品统计

```python
df1 = df
# df1.groupby('商品名称', as_index = False)
df1.sort_values('销售数量', inplace = True, ascending = False)
df1.groupby('商品名称', as_index = False)['销售数量'].sum()
df1 = df1.head(10)
x = df1.商品名称
y = df1.销售数量
plt.figure(figsize = (20, 4))
plt.rc('font', **{'family': 'SimHei'})
plt.bar(x, y)
plt.show()
```

#### 日销量统计

```python
df1 = df
df1 = df1.groupby('销售时间', as_index = False)['销售数量'].sum()
df1.sort_values('销售数量', inplace = True, ascending = False)
print(df1)
df1 = df1.head(10)
x = df1.销售时间
y = df1.销售数量
plt.figure(figsize = (10, 4))
plt.rc('font', **{'family': 'SimHei'})
plt.bar(x, y)
plt.show()
```
