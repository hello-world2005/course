```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('超市销售数据.csv', encoding = 'GBK')
```

#### 最受欢迎的商品统计

```python
df1 = df
# df1.groupby('商品名称', as_index = False)
df1.sort_values('销售数量', inplace = True, ascending = False)
df2 = df1.groupby('商品名称', as_index = False)['销售数量'].sum()
df2.sort_values('销售数量', inplace = True, ascending = False)
df2 = df2.head(10)
x = df2.商品名称
y = df2.销售数量
plt.figure(figsize = (20, 4))
plt.rc('font', **{'family': 'SimHei'})
plt.bar(x, y)
plt.show()
```

#### 潜在最大客户（雾

```python
df1 = df
df2 = df1.groupby('单据号', as_index = False)['销售金额'].sum()
df2.sort_values('销售金额', inplace = True, ascending = False)
df2 = df2.head(10)
print(df2)
x = df2.单据号
y = df2.销售金额
plt.figure(figsize = (20, 4))
plt.rc('font', **{'family': 'SimHei'})
plt.bar(x, y)
plt.show()
```
